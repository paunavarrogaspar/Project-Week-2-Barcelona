```python
#Importing the Data Frame that includes information about air quality
pollution=pd.read_csv("../datasets/2.-Urban-Environment/air_quality_2018_project2.csv", index_col="Unnamed: 0")
pollution.head()
```

```python
#Cleaning the pollution dataset. We are taking the Barcelona - info from the neighborhoods to have it in the same format as the info about neighborhoods in our public transportation dataframe
#Taking the unit from the numerical columns to make then a int datatype instead of an object datatype
pollution["nom_cabina"]=pollution["nom_cabina"].str.replace(".*(- )","")
pollution["valor_o3"]=pollution["valor_o3"].str.replace(" µg/m³","")
pollution["valor_no2"]=pollution["valor_no2"].str.replace(" µg/m³","")
pollution["valor_pm10"]=pollution["valor_pm10"].str.replace(" µg/m³","")
# test transformed data
pollution.columns
```
